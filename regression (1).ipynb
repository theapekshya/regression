```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Salary Data.csv',usecols = ['Years of Experience','Salary'])
dataset = dataset.dropna()
dataset
```

```python
 #dataset[['Years of Experience']].values
```

```python
#dataset['Years of Experience']
```

```python
X = dataset.iloc[:, 0:-1].values
X
```

```python
Y = dataset.iloc[:,1:].values
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.1,random_state = 10)
```

```python
print(X_train)
```

```python
total_len = (len(X))
print("All total length = ",total_len)
print("X_train length = ", len(X_train),len(Y_train))
print("X_test length = ", len(X_test))
```

```python
#Training model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
#testing model
exp = float(input("Enter years of experience = "))
exp = np.array([[exp]])
print(exp)

# # predict the salary
predicted_salary = model.predict(exp)[0][0]
# print(predicted_salary)
print("Your predicted_salary = ",predicted_salary)
```
